# Lexical access

plans a question by looking up the semantic annotation in human-human data

```python
from emo20q.lexicalaccess import LexicalAccess
la = LexicalAccess()
```

```python
# by default, get one random example of the given semantic annotation
la.lookUp('directed(e,otherPerson)')
```

```python
# but to access all the options, one can access the _dictionary object
for realization in la._dictionary['directed(e,otherPerson)']:
    print(realization)
```

```python
# here are all the semantic annotations, with an example realization
for annotation in la._dictionary:
    print(f"annotation: {annotation}, \n\trealization: {la.lookUp(annotation)}")
```
